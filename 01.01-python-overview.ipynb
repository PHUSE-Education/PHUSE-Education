# Overview
Python is an open source, object oriented, multi-purpose programming language, which can be used for anything including clinical data science. This makes Python a very strong language since clinical data science practitioners need a wide set of skills like data mining, data analyzing, visualizing and building machine learning models. Python is widely used in the scientific and research communities. It is easy to use thanks to its simple syntax. Python offers great libraries, such as Numpy for complex mathematical derivations, Pandas for efficient data handling, or Matplotlib for visualization. These well documented libraries make data science tasks a lot easier.

Dating from 1991, the Python programming language was considered a gap-filler, a way to write scripts that automate the boring stuff (as one popular book on learning Python put it) or to rapidly prototype applications that will be implemented in other languages.  However, over the past few years, Python has emerged as a first-class citizen in modern software development, infrastructure management, and data analysis. It is no longer a back-room utility language, but a major force in web application, creation and systems management, and a key driver of the explosion in big data analytics and machine intelligence.


# Getting Started with Python
The first low hurdle to take if you want to start Python is to get access to Python. Since Python is an open source programming language, there are multiple ways to get access. The following describes how you as an individual user can start working with Python either on your local machine or in the cloud. If you want to utilize Python in a corporate setting or even want to deploy a validated Python instance in a validated clinical environment, please involve your IT department.

Probably the easiest way how to get started with Python is using one of the many available cloud options. Often, these options do not even include the requirement to create an account. One very nice example is provided by the Jupyter project community.

The open source platform Jupyter provides access to Jupyter Notebooks in your browser. There are options to install Jupyter Notebooks as well, but we will share later easier options to install Jupyter notebooks on your machine.

If you want to try it in your browser, go to www.jupyter.org and click on the Try it on your browser button. The next screen will offer you the option to select the programming language of your choice. As you can see, you can select from multiple options besides Python like R, Julia, C++ and more. Juypter actually supports more than 40 programming languages. Clicking on any programming language will start a temporary virtual server and opens a Jupyter notebook with further explanations.

Since we want to start with Python, simply click on Try Classic Notebook, which will start a virtual server and open a Jupyter Notebook using a Python 3 kernel.
